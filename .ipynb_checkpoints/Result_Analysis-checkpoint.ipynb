```python
import os
import numpy as np
import pandas as pd
from scipy.stats import pearsonr

import geopandas as gpd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as mticker
import matplotlib.patches as patches
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.ticker import MaxNLocator
from matplotlib.gridspec import GridSpec

from datetime import timedelta
from sklearn.metrics import r2_score, mean_absolute_error

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('final.csv')
shapefile = gpd.read_file(r"watershed")
```

```python
df.head(5)
```

```python
df.drop(columns=['Unnamed: 0', 'New_Prediction'], inplace=True)
df=df[['Date','Lat', 'Lon', 'Lead_Time', 'Observed_Precip', 'Ensemble_Mean', 'Distilled_Prediction']]
df['Date']=pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.head(5)
```

```python
df = df[np.isfinite(df[["Observed_Precip", "Distilled_Prediction"]].values).all(axis=1)].copy()
df['Distilled_Prediction'] = df['Distilled_Prediction'].clip(lower=0)

grouped = df.groupby(['Lat', 'Lon', 'Lead_Time'])[['Observed_Precip', 'Ensemble_Mean', 'Distilled_Prediction']].sum().reset_index()

grouped['Model_Bias'] = grouped['Ensemble_Mean'] - grouped['Observed_Precip']
grouped['Distilled_Bias'] = grouped['Distilled_Prediction'] - grouped['Observed_Precip']
```

```python
# Total Rainfall Percentage Bias Maps
```

```python
def plot_total_bias_percentage_maps(data, shapefile, leads=[1, 2, 3, 4, 5], vmin=-80, vmax=80, cmap='bwr'):
    degree_formatter = mticker.FuncFormatter(lambda x, pos: f"{int(x)}°")

    bounds = np.linspace(vmin, vmax, 17)
    cmap_discrete = plt.get_cmap(cmap, len(bounds) - 1)
    norm = BoundaryNorm(bounds, cmap_discrete.N)

    ncols = len(leads)
    fig = plt.figure(figsize=(5 * ncols, 6))

    left_margin = 0.06
    right_margin = 0.97
    top_margin = 0.95
    bottom_margin = 0.12
    h_gap = 0.02
    v_gap = 0.005

    subplot_width = (right_margin - left_margin - (ncols - 1) * h_gap) / ncols
    subplot_height = 0.38

    last_im = None

    for col_idx, lead_day in enumerate(leads):
        subset = data[data['Lead_Time'] == lead_day].copy()
        subset['Model_Bias'] = 100 * subset['Model_Bias'] / subset['Observed_Precip'].replace(0, np.nan)
        subset['Distilled_Bias'] = 100 * subset['Distilled_Bias'] / subset['Observed_Precip'].replace(0, np.nan)

        for row_idx, var in enumerate(['Model_Bias', 'Distilled_Bias']):
            left = left_margin + col_idx * (subplot_width + h_gap)
            bottom = top_margin - (row_idx + 1) * (subplot_height + v_gap)

            ax = fig.add_axes([left, bottom, subplot_width, subplot_height])

            grid = subset.pivot_table(index='Lat', columns='Lon', values=var, fill_value=np.nan).sort_index(ascending=False)
            extent = [grid.columns.min(), grid.columns.max(), grid.index.min(), grid.index.max()]

            shapefile.boundary.plot(ax=ax, linewidth=1.2, edgecolor='black', alpha=0.9)
            im = ax.imshow(grid.values, extent=extent, origin='upper', cmap=cmap_discrete, norm=norm)
            last_im = im

            rect = patches.Rectangle((0, 0), 1, 1, transform=ax.transAxes,
                                     fill=False, color='black', linewidth=2)
            ax.add_patch(rect)

            if row_idx == 0:
                ax.set_title(f"{lead_day} Day Lead", fontsize=17, weight='bold', pad=10)

            if col_idx == 0:
                ax.set_ylabel('Ensemble Bias' if row_idx == 0 else 'Corrected Bias', fontsize=15, weight='bold')
                ax.set_yticks([20, 21, 22, 23, 24])
                ax.set_yticklabels([f"{y}°" for y in [20, 21, 22, 23, 24]])
                ax.tick_params(axis='y', labelsize=16, width=2, length=6)
            else:
                ax.set_yticklabels([])

            if row_idx == 1:
                ax.set_xticks([73, 75, 77, 79, 81])
                ax.set_xticklabels([f"{x}°" for x in [73, 75, 77, 79, 81]])
                ax.tick_params(axis='x', labelsize=16, width=2, length=6)
            else:
                ax.set_xticklabels([])

            ax.set_xlim(73, 81.875)
            ax.set_ylim(20, 24.375)

    cbar_left = left_margin + 0.15
    cbar_width = (right_margin - left_margin) - 0.3
    cbar_bottom = bottom_margin - 0.075
    cbar_ax = fig.add_axes([cbar_left, cbar_bottom, cbar_width, 0.02])
    cbar = fig.colorbar(last_im, cax=cbar_ax, orientation='horizontal', ticks=bounds, norm=norm)
    cbar.set_ticklabels([str(int(t)) for t in bounds])
    cbar.set_label('Percentage Bias (%)', fontsize=18, weight='bold')
    cbar.ax.tick_params(labelsize=16, width=2, length=6)

    save_path = r"C:\Users\manas\Downloads\_srip\results\bias_percentage_maps.png"
    plt.savefig(save_path, dpi=600, bbox_inches='tight')
    print(f"Figure saved to: {save_path}")

    plt.show()
```

```python
plot_total_bias_percentage_maps(grouped, shapefile, leads=[1, 2, 3, 4, 5])
```

```python
def compute_extreme_bias_percentage(group):
    if len(group) < 10:
        return pd.Series({'Model_Bias_extreme': np.nan, 'Distilled_Bias_extreme': np.nan})

    threshold = group['Observed_Precip'].quantile(0.9)
    extreme_group = group[group['Observed_Precip'] > threshold]

    if len(extreme_group) < 1:
        return pd.Series({'Model_Bias_extreme': np.nan, 'Distilled_Bias_extreme': np.nan})

    obs_sum = extreme_group['Observed_Precip'].sum()
    ens_sum = extreme_group['Ensemble_Mean'].sum()
    distill_sum = extreme_group['Distilled_Prediction'].sum()

    return pd.Series({
        'Model_Bias_extreme': 100 * (ens_sum - obs_sum) / obs_sum,
        'Distilled_Bias_extreme': 100 * (distill_sum - obs_sum) / obs_sum
    })

extreme_bias_df = df.groupby(['Lat', 'Lon', 'Lead_Time']).apply(compute_extreme_bias_percentage).reset_index()
```

```python
def plot_extreme_bias_maps(data, shapefile, leads=[1, 2, 3, 4, 5], vmin=-80, vmax=80, cmap='bwr'):
    variables = [
        ('Model_Bias_extreme', 'Ensemble Bias'),
        ('Distilled_Bias_extreme', 'Corrected Bias')
    ]

    # Define discrete bins and colormap
    bounds = np.arange(vmin, vmax + 10, 10)
    cmap_discrete = plt.get_cmap(cmap, len(bounds) - 1)
    norm = BoundaryNorm(bounds, cmap_discrete.N)

    degree_formatter = mticker.FuncFormatter(lambda x, pos: f"{int(x)}°")

    ncols = len(leads)
    fig = plt.figure(figsize=(5 * ncols, 6))

    left_margin = 0.06
    right_margin = 0.97
    top_margin = 0.95
    bottom_margin = 0.12
    h_gap = 0.02
    v_gap = 0.005

    subplot_width = (right_margin - left_margin - (ncols - 1) * h_gap) / ncols
    subplot_height = 0.38

    last_im = None

    for col_idx, lead_day in enumerate(leads):
        subset = data[data['Lead_Time'] == lead_day].copy()

        for row_idx, (var, row_title) in enumerate(variables):
            left = left_margin + col_idx * (subplot_width + h_gap)
            bottom = top_margin - (row_idx + 1) * (subplot_height + v_gap)

            ax = fig.add_axes([left, bottom, subplot_width, subplot_height])

            grid = subset.pivot_table(index='Lat', columns='Lon', values=var, fill_value=np.nan).sort_index(ascending=False)
            extent = [grid.columns.min(), grid.columns.max(), grid.index.min(), grid.index.max()]

            shapefile.boundary.plot(ax=ax, linewidth=1.2, edgecolor='black', alpha=0.9)
            im = ax.imshow(grid.values, extent=extent, origin='upper', cmap=cmap_discrete, norm=norm)
            last_im = im

            rect = patches.Rectangle((0, 0), 1, 1, transform=ax.transAxes,
                                     fill=False, color='black', linewidth=2)
            ax.add_patch(rect)

            if row_idx == 0:
                ax.set_title(f"{lead_day} Day Lead", fontsize=17, weight='bold', pad=10)

            if col_idx == 0:
                ax.set_ylabel(row_title, fontsize=15, weight='bold')
                ax.set_yticks([20, 21, 22, 23, 24])
                ax.set_yticklabels([f"{y}°" for y in [20, 21, 22, 23, 24]])
                ax.tick_params(axis='y', labelsize=16, width=2, length=6)
            else:
                ax.set_yticklabels([])

            if row_idx == 1:
                ax.set_xticks([73, 75, 77, 79, 81])
                ax.set_xticklabels([f"{x}°" for x in [73, 75, 77, 79, 81]])
                ax.tick_params(axis='x', labelsize=16, width=2, length=6)
            else:
                ax.set_xticklabels([])

            ax.set_xlim(73, 81.875)
            ax.set_ylim(20, 24.375)

    # Colorbar
    cbar_left = left_margin + 0.15
    cbar_width = (right_margin - left_margin) - 0.3
    cbar_bottom = bottom_margin - 0.075
    cbar_ax = fig.add_axes([cbar_left, cbar_bottom, cbar_width, 0.02])
    cbar = fig.colorbar(last_im, cax=cbar_ax, orientation='horizontal', ticks=bounds, norm=norm)
    cbar.set_ticklabels([str(int(t)) for t in bounds])
    cbar.set_label('Extreme Percentage Bias (%)', fontsize=18, weight='bold')
    cbar.ax.tick_params(labelsize=16, width=2, length=6)

    # Save
    save_path = r"C:\Users\manas\Downloads\_srip\results\extreme_bias_percentage_maps.png"
    plt.savefig(save_path, dpi=600, bbox_inches='tight')
    print(f"Figure saved to: {save_path}")

    plt.show()
```

```python
def compute_correlations(group):
    if len(group) < 3:
        return pd.Series({'Corr_Ensemble': np.nan, 'Corr_Distilled': np.nan})
    corr1 = pearsonr(group['Observed_Precip'], group['Ensemble_Mean'])[0]
    corr2 = pearsonr(group['Observed_Precip'], group['Distilled_Prediction'])[0]
    return pd.Series({'Corr_Ensemble': corr1, 'Corr_Distilled': corr2})

correlation_df = df.groupby(['Lat', 'Lon', 'Lead_Time']).apply(compute_correlations).reset_index()
```

```python
def compute_total_nrmse(group):
    if len(group) < 3 or group['Observed_Precip'].mean() == 0:
        return pd.Series({'NRMSE_Model': np.nan, 'NRMSE_Distilled': np.nan})

    rmse_model = np.sqrt(np.mean((group['Ensemble_Mean'] - group['Observed_Precip']) ** 2))
    rmse_distill = np.sqrt(np.mean((group['Distilled_Prediction'] - group['Observed_Precip']) ** 2))
    mean_obs = group['Observed_Precip'].mean()

    return pd.Series({
        'NRMSE_Model': rmse_model / mean_obs,
        'NRMSE_Distilled': rmse_distill / mean_obs
    })

nrmse_df = df.groupby(['Lat', 'Lon', 'Lead_Time']).apply(compute_total_nrmse).reset_index()
```

```python
def compute_extreme_nrmse(group):
    if len(group) < 10:
        return pd.Series({'NRMSE_Model_extreme': np.nan, 'NRMSE_Distilled_extreme': np.nan})

    threshold = group['Observed_Precip'].quantile(0.9)
    extreme_group = group[group['Observed_Precip'] > threshold]

    if len(extreme_group) < 3 or extreme_group['Observed_Precip'].mean() == 0:
        return pd.Series({'NRMSE_Model_extreme': np.nan, 'NRMSE_Distilled_extreme': np.nan})

    rmse_model = np.sqrt(np.mean((extreme_group['Ensemble_Mean'] - extreme_group['Observed_Precip']) ** 2))
    rmse_distill = np.sqrt(np.mean((extreme_group['Distilled_Prediction'] - extreme_group['Observed_Precip']) ** 2))
    mean_obs = extreme_group['Observed_Precip'].mean()

    return pd.Series({
        'NRMSE_Model_extreme': rmse_model / mean_obs,
        'NRMSE_Distilled_extreme': rmse_distill / mean_obs
    })

nrmse_extreme_df = df.groupby(['Lat', 'Lon', 'Lead_Time']).apply(compute_extreme_nrmse).reset_index()
```

```python
def plot_extreme_precip_maps(data, shapefile, leads=[1, 2, 3, 4, 5], vmin=0, vmax=250, cmap='Blues'):
    variables = [
        ('Observed_Precip', 'Observed'),
        ('Ensemble_Mean', 'Ensemble Model'),
        ('Distilled_Prediction', 'New Model'),
    ]

    bounds = np.linspace(vmin, vmax, 11)
    cmap_discrete = plt.get_cmap(cmap, len(bounds) - 1)
    norm = BoundaryNorm(bounds, cmap_discrete.N)

    extreme_data = []
    grouped = data.groupby(['Lat', 'Lon', 'Lead_Time'])

    for (lat, lon, lead), group in grouped:
        if len(group) < 10:
            continue
        q90 = group['Observed_Precip'].quantile(0.9)
        extreme = group[group['Observed_Precip'] > q90]
        if not extreme.empty:
            agg = extreme[['Observed_Precip', 'Ensemble_Mean', 'Distilled_Prediction']].sum()
            agg['Lat'] = lat
            agg['Lon'] = lon
            agg['Lead_Time'] = lead
            extreme_data.append(agg)

    extreme_df = pd.DataFrame(extreme_data)

    ncols = len(leads)
    nrows = len(variables)
    fig = plt.figure(figsize=(5 * ncols, 7))

    left_margin = 0.06
    right_margin = 0.97
    top_margin = 0.95
    bottom_margin = 0.18

    h_gap = 0.005
    v_gap = 0.07

    subplot_width = (right_margin - left_margin - (ncols - 1) * h_gap) / ncols
    subplot_height = (top_margin - bottom_margin - (nrows - 1) * v_gap) / nrows

    last_im = None
    for col_idx, lead_day in enumerate(leads):
        subset = extreme_df[extreme_df['Lead_Time'] == lead_day]

        for row_idx, (var, row_title) in enumerate(variables):
            left = left_margin + col_idx * (subplot_width + h_gap)
            bottom = top_margin - (row_idx + 1) * (subplot_height + v_gap)

            ax = fig.add_axes([left, bottom, subplot_width, subplot_height])

            grid = subset.pivot_table(index='Lat', columns='Lon', values=var, fill_value=np.nan).sort_index(ascending=False)
            extent = [grid.columns.min(), grid.columns.max(), grid.index.min(), grid.index.max()]

            shapefile.boundary.plot(ax=ax, linewidth=1.2, edgecolor='black', alpha=0.9)
            im = ax.imshow(grid.values, extent=extent, origin='upper', cmap=cmap_discrete, norm=norm)
            last_im = im

            rect = patches.Rectangle((0, 0), 1, 1, transform=ax.transAxes, fill=False, color='black', linewidth=2)
            ax.add_patch(rect)

            if row_idx == 0:
                ax.set_title(f"{lead_day} Day Lead", fontsize=17, weight='bold', pad=10)

            if col_idx == 0:
                ax.set_ylabel(row_title, fontsize=15, weight='bold')
                ax.set_yticks([20, 21, 22, 23, 24])
                ax.set_yticklabels([f"{y}°" for y in [20, 21, 22, 23, 24]])
                ax.tick_params(axis='y', labelsize=16, width=2, length=6)
            else:
                ax.set_yticklabels([])

            ax.set_xticks([73, 75, 77, 79, 81])
            ax.set_xticklabels([f"{x}°" for x in [73, 75, 77, 79, 81]])
            ax.tick_params(axis='x', labelsize=16, width=2, length=6)

            ax.set_xlim(73, 81.875)
            ax.set_ylim(20, 24.375)

    cbar_left = left_margin + 0.1
    cbar_width = (right_margin - left_margin) - 0.2
    cbar_bottom = bottom_margin - 0.23
    cbar_ax = fig.add_axes([cbar_left, cbar_bottom, cbar_width, 0.025])
    cbar = fig.colorbar(last_im, cax=cbar_ax, orientation='horizontal', ticks=bounds, norm=norm)
    cbar.set_ticklabels([f"{t:.0f}" for t in bounds])
    cbar.set_label('Extreme Total Precipitation (mm)', fontsize=18, weight='bold')
    cbar.ax.tick_params(labelsize=16, width=2, length=6)

    save_path = r"C:\Users\manas\Downloads\_srip\results\extreme_precip_maps.png"
    plt.savefig(save_path, dpi=600, bbox_inches='tight')
    print(f"Figure saved to: {save_path}")
    plt.show()
```

```python
def plot_max_event_layout(data, shapefile, lead_day_max_dates, vmin_p=0, vmax_p=100, vmin_b=-80, vmax_b=80):
    fig = plt.figure(figsize=(18, 10))

    ncols = 5
    nrows = 3
    left_margin = 0.06
    right_margin = 0.97
    top_margin = 0.93
    bottom_margin = 0.12
    h_gap = 0.015
    v_gap = 0.03

    subplot_width = (right_margin - left_margin - (ncols - 1) * h_gap) / ncols
    subplot_height = (top_margin - bottom_margin - (nrows - 1) * v_gap) / nrows

    lead_days = [1, 2, 3, 4, 5]

    bounds_p = np.linspace(vmin_p, vmax_p, 11)
    cmap_p = plt.get_cmap('Blues', len(bounds_p) - 1)
    norm_p = BoundaryNorm(bounds_p, cmap_p.N)

    bounds_b = np.linspace(vmin_b, vmax_b, 17)
    cmap_b = plt.get_cmap('bwr', len(bounds_b) - 1)
    norm_b = BoundaryNorm(bounds_b, cmap_b.N)

    cbar_data = {'Observed_Precip': [], 'Model_Bias': [], 'Distilled_Bias': []}

    for col_idx, lead_day in enumerate(lead_days):
        date = lead_day_max_dates.get(lead_day)
        if date is None:
            continue

        subset = data[(data.index == date) & (data['Lead_Time'] == lead_day)].copy()
        subset['Model_Bias'] = 100 * (subset['Ensemble_Mean'] - subset['Observed_Precip']) / subset['Observed_Precip'].replace(0, np.nan)
        subset['Distilled_Bias'] = 100 * (subset['Distilled_Prediction'] - subset['Observed_Precip']) / subset['Observed_Precip'].replace(0, np.nan)

        variables = [
            ('Observed_Precip', cmap_p, norm_p, bounds_p),
            ('Model_Bias', cmap_b, norm_b, bounds_b),
            ('Distilled_Bias', cmap_b, norm_b, bounds_b),
        ]

        for row_idx, (var, cmap, norm, _) in enumerate(variables):
            left = left_margin + col_idx * (subplot_width + h_gap)
            bottom = top_margin - (row_idx + 1) * (subplot_height + v_gap * row_idx)

            ax = fig.add_axes([left, bottom, subplot_width, subplot_height])
            grid = subset.pivot_table(index='Lat', columns='Lon', values=var, fill_value=np.nan).sort_index(ascending=False)
            extent = [grid.columns.min(), grid.columns.max(), grid.index.min(), grid.index.max()]
            shapefile.boundary.plot(ax=ax, linewidth=1.2, edgecolor='black', alpha=0.9)
            im = ax.imshow(grid.values, extent=extent, origin='upper', cmap=cmap, norm=norm)

            rect = patches.Rectangle((0, 0), 1, 1, transform=ax.transAxes, fill=False, color='black', linewidth=2)
            ax.add_patch(rect)

            ax.set_xticks([73, 75, 77, 79, 81])
            ax.set_xticklabels([f"{x}°" for x in [73, 75, 77, 79, 81]])
            ax.tick_params(axis='x', labelsize=12, width=1.5)

            if col_idx == 0:
                ax.set_yticks([20, 21, 22, 23, 24])
                ax.set_yticklabels([f"{y}°" for y in [20, 21, 22, 23, 24]])
                ax.tick_params(axis='y', labelsize=12, width=1.5)
            else:
                ax.set_yticklabels([])

            ax.set_xlim(73, 81.875)
            ax.set_ylim(20, 24.375)

            cbar_data[var].append(im)
            
    row_specs = [
        ('Observed_Precip', 'Observed Rainfall (mm)', bounds_p),
        ('Model_Bias', 'Ensemble Model Bias (%)', bounds_b),
        ('Distilled_Bias', 'New Model Bias (%)', bounds_b)
    ]

    for row_idx, (var, label, bounds) in enumerate(row_specs):
        cbar_left = left_margin + 0.1
        cbar_width = (right_margin - left_margin) - 0.2
        cbar_bottom = top_margin - (row_idx + 1) * (subplot_height + v_gap * row_idx) - 0.035
        cbar_ax = fig.add_axes([cbar_left, cbar_bottom, cbar_width, 0.015])
        cbar = fig.colorbar(cbar_data[var][2], cax=cbar_ax, orientation='horizontal', ticks=bounds)
        cbar.set_ticklabels([str(int(t)) for t in bounds])
        cbar.set_label(label, fontsize=15)
        cbar.ax.tick_params(labelsize=11, width=1.5)

    for col_idx, lead_day in enumerate(lead_days):
        left = left_margin + col_idx * (subplot_width + h_gap) + subplot_width / 2
        y_pos = top_margin + 0.005
        fig.text(left, y_pos, f"Lead Day {lead_day}", ha='center', va='bottom', fontsize=14, weight='bold')

    row_titles = ["Observed Rainfall", "Ensemble Model Bias", "New Model Bias"]
    for row_idx, title in enumerate(row_titles):
        y_center = top_margin - (row_idx + 0.5) * (subplot_height + v_gap * row_idx)
        fig.text(left_margin - 0.035, y_center, title, ha='right', va='center',
                 fontsize=14, weight='bold', rotation=90)

    save_path = r"C:\Users\manas\Downloads\_srip\results\max_flood_event_grid_final.png"
    plt.savefig(save_path, dpi=600, bbox_inches='tight')
    print(f"Saved: {save_path}")
    plt.show()
```

```python
df.index = pd.to_datetime(df.index)
df = df.sort_index()

fig = plt.figure(figsize=(15, 12))
gs = GridSpec(3, 2, figure=fig)

axes = []
for i in range(5):
    if i < 4:
        ax = fig.add_subplot(gs[i // 2, i % 2])
    else:
        ax = fig.add_subplot(gs[2, :])
    axes.append(ax)

plot_count = 0

for lead in range(1, 6):
    window_data = []
    date_refs = []
    max_rain_dates = []

    for (lat, lon), group in df[df['Lead_Time'] == lead].groupby(['Lat', 'Lon']):
        if group.empty:
            continue

        max_rain_date = group['Observed_Precip'].idxmax()
        release_day = max_rain_date - timedelta(days=lead)
        window_start = release_day - timedelta(days=3)
        window_end = release_day + timedelta(days=6)

        grid_window = df.loc[window_start:window_end]
        grid_window = grid_window[(grid_window['Lat'] == lat) & (grid_window['Lon'] == lon)]

        if len(grid_window) == 10:
            window_data.append(grid_window[['Observed_Precip', 'Ensemble_Mean', 'Distilled_Prediction']].reset_index(drop=True))
            date_refs.append(grid_window.index)
            max_rain_dates.append(max_rain_date)

    if not window_data:
        continue

    combined = pd.concat(window_data, axis=1)
    avg_series = combined.groupby(level=0, axis=1).mean()

    date_index = date_refs[0]
    max_rain_date = max_rain_dates[0]
    forecast_date = max_rain_date - timedelta(days=lead)

    before_idx = date_index <= forecast_date
    after_idx = date_index >= forecast_date

    ax = axes[plot_count]
    plot_count += 1

    ax.plot(date_index[before_idx], avg_series['Observed_Precip'][before_idx],
             color='black', linewidth=2.5, label='Observed')
    ax.plot(date_index[after_idx], avg_series['Observed_Precip'][after_idx],
             color='black', linewidth=2.5)

    ax.plot(date_index[after_idx], avg_series['Ensemble_Mean'][after_idx],
             color='#d62728', linewidth=2.2, label='Ensemble Mean')
    ax.plot(date_index[after_idx], avg_series['Distilled_Prediction'][after_idx],
             color='#2ca02c', linewidth=2.2, label='Distilled')

    ax.axvline(forecast_date, color='blue', linestyle='--', linewidth=1.8, label='Forecast Release')
    ax.axvline(max_rain_date, color='purple', linestyle='--', linewidth=1.8, label='Max Rainfall Day')

    ax.set_title(f'Lead {lead} - Maximum Rainfall: {max_rain_date.date()}', fontsize=13)
    ax.set_xlabel('Date', fontsize=11)
    ax.set_ylabel('Rainfall (mm)', fontsize=11)
    ax.tick_params(axis='both', labelsize=10)
    ax.grid(True, linestyle=':', linewidth=0.7)

    ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d, %Y'))
    ax.xaxis.set_major_locator(MaxNLocator(nbins=6))
    ax.tick_params(axis='x', rotation=0)

    if lead in [1]:
        ax.legend(fontsize=9, frameon=True, framealpha=0.9)

plt.tight_layout()
plt.savefig("forecast_plots.png", dpi=600)
plt.show()
```

```python
def nse(obs, sim):
    return 1 - np.sum((obs - sim) ** 2) / np.sum((obs - np.mean(obs)) ** 2)

def pbias(obs, sim):
    return 100 * (np.sum(sim - obs) / np.sum(obs)) if np.sum(obs) != 0 else np.nan

def nrmse(obs, sim):
    rmse = np.sqrt(np.mean((obs - sim) ** 2))
    return rmse / np.mean(obs) if np.mean(obs) != 0 else np.nan

def kge(obs, sim):
    r = np.corrcoef(obs, sim)[0, 1]
    alpha = np.std(sim) / np.std(obs) if np.std(obs) != 0 else np.nan
    beta = np.mean(sim) / np.mean(obs) if np.mean(obs) != 0 else np.nan
    return 1 - np.sqrt((r - 1) ** 2 + (alpha - 1) ** 2 + (beta - 1) ** 2)

def compute_metrics(obs, sim):
    if len(obs) == 0 or np.all(np.isnan(obs)) or np.all(np.isnan(sim)):
        return [np.nan]*6
    return [
        r2_score(obs, sim),
        nse(obs, sim),
        kge(obs, sim),
        pbias(obs, sim),
        mean_absolute_error(obs, sim),
        nrmse(obs, sim)
    ]
```

```python
lead_times = [1, 2, 3, 4, 5]
x_labels = ["Overall"] + [f"Lead {lt}" for lt in lead_times]
metrics_labels = ["R²", "NSE", "KGE", "PBIAS (%)", "MAE (mm)", "nRMSE"]
ensemble_scores = [[] for _ in range(6)]
distilled_scores = [[] for _ in range(6)]

obs = df["Observed_Precip"]
ens = df["Ensemble_Mean"]
new = df["Distilled_Prediction"]

for i, m in enumerate(compute_metrics(obs, ens)):
    ensemble_scores[i].append(m)
for i, m in enumerate(compute_metrics(obs, new)):
    distilled_scores[i].append(m)

for lt in lead_times:
    subset = df[df["Lead_Time"] == lt]
    obs = subset["Observed_Precip"]
    ens = subset["Ensemble_Mean"]
    new = subset["Distilled_Prediction"]
    
    for i, m in enumerate(compute_metrics(obs, ens)):
        ensemble_scores[i].append(m)
    for i, m in enumerate(compute_metrics(obs, new)):
        distilled_scores[i].append(m)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

x = np.arange(len(x_labels))
width = 0.35

for i in range(6):
    ax = axes[i]
    ax.bar(x - width/2, ensemble_scores[i], width, label="Ensemble Model", color="#1f77b4")
    ax.bar(x + width/2, distilled_scores[i], width, label="New Model", color="#2ca02c")

    for j in range(len(x_labels)):
        if not np.isnan(ensemble_scores[i][j]):
            ax.text(x[j] - width/2, ensemble_scores[i][j] + 0.01, f"{ensemble_scores[i][j]:.2f}", 
                    ha='center', va='bottom', fontsize=8)
        if not np.isnan(distilled_scores[i][j]):
            ax.text(x[j] + width/2, distilled_scores[i][j] + 0.01, f"{distilled_scores[i][j]:.2f}", 
                    ha='center', va='bottom', fontsize=8)
    
    ax.set_title(metrics_labels[i], fontsize=12, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels(x_labels)
    ax.set_ylabel(metrics_labels[i])
    ax.grid(True, linestyle='--', alpha=0.5)

    if i == 0:
        ax.legend()
    
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    ax.spines['left'].set_visible(True)
    ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("bias_correction_metrics_comparison.png", dpi=600, bbox_inches='tight')
plt.show()
```
